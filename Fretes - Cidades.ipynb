```python
navegador = webdriver.Chrome()
navegador.get('https://erp.sincsuite.com.br')
time.sleep(2)
login = navegador.find_element(By.ID, 'txtUser').send_keys('cayo.garcia' + Keys.TAB)
senha = navegador.find_element(By.ID, 'txtPwd').send_keys('Flavia01@#')
navegador.find_element(By.XPATH, '//*[@id="btnLogin"]').click()
time.sleep(3)

navegador.find_element(By.XPATH, '//*[@id="visionTitle54"]').click()
time.sleep(2)
navegador.find_element(By.XPATH, '//*[@id="panelTitle58"]').click()
time.sleep(1)
iframe = navegador.find_element(By.XPATH, '//*[@id="frmPrincipal"]')
navegador.switch_to.frame(iframe)
time.sleep(2)
for index, linha in tabela.iterrows():
    navegador.find_element(By.XPATH, '/html/body/div[3]/div/table/tbody/tr/td[1]/div/a[1]').click()
    time.sleep(2)
    depositos = navegador.find_element(By.XPATH, '//*[@id="modalPanel_modalDialog"]/div/div[2]/table[1]/tbody/tr[1]/td[3]/div/button').click()
    selecao = navegador.find_element(By.XPATH, '//*[@id="modalPanel_modalDialog"]/div/div[2]/table[1]/tbody/tr[1]/td[3]/div/div/div[1]/input')
    selecao.send_keys('P. MARIA TERESA - MOGI' + Keys.ENTER)
    tipo_frete = navegador.find_element(By.XPATH, '//*[@id="ddlInd_FreightType"]')
    tipo_frota = navegador.find_element(By.XPATH, '//*[@id="modalPanel_modalDialog"]/div/div[2]/table[2]/tbody/tr/td[3]/div/button').click()
    selecao_frota = navegador.find_element(By.XPATH, '//*[@id="modalPanel_modalDialog"]/div/div[2]/table[2]/tbody/tr/td[3]/div/div/div[1]/input')
    selecao_frota.send_keys(linha["ESPÉCIE"] + Keys.ENTER)
    unidade = navegador.find_element(By.XPATH, '//*[@id="modalPanel_modalDialog"]/div/div[2]/table[2]/tbody/tr/td[5]/div/button').click()
    selecao_unidade = navegador.find_element(By.XPATH, '//*[@id="modalPanel_modalDialog"]/div/div[2]/table[2]/tbody/tr/td[5]/div/div/div[1]/input')
    selecao_unidade.send_keys(linha["UNIDADE"] + Keys.ENTER)
    valor = navegador.find_element(By.XPATH, '//*[@id="txtVal_FreightValue"]').send_keys(str(linha["VALOR"]))
    estado = navegador.find_element(By.XPATH, '//*[@id="panelFCB"]/table/tbody/tr/td[1]/div/button').click()
    selecao_estado = navegador.find_element(By.XPATH, '//*[@id="panelFCB"]/table/tbody/tr/td[1]/div/div/div[1]/input')
    selecao_estado.send_keys('São Paulo' + Keys.ENTER)
    time.sleep(1)
    cidade = navegador.find_element(By.XPATH, '//*[@id="panelFCB"]/table/tbody/tr/td[3]/div/button').click()
    time.sleep(1)
    cidade_selecao = navegador.find_element(By.XPATH, '//*[@id="panelFCB"]/table/tbody/tr/td[3]/div/div/div[1]/input')
    cidade_selecao.send_keys(linha["CIDADE"])
    time.sleep(1)
    cidade_selecao.send_keys(Keys.ENTER)
    navegador.find_element(By.XPATH, '//*[@id="modalPanelSaveBtn"]').click()
    time.sleep(3)

navegador.quit()
```
